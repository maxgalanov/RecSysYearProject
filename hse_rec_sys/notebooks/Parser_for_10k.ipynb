```python
def get_info_h5(h5):
    artist_name = get_artist_name(h5).decode('UTF-8')
    artist_familarity = float(get_artist_familiarity(h5))
    artist_hotttnesss = float(get_artist_hotttnesss(h5))
    artist_id = get_artist_id(h5).decode('UTF-8')
    artist_location = get_artist_location(h5).decode('UTF-8')
    danceability = float(get_danceability(h5))
    duration = float(get_duration(h5))
    energy = float(get_energy(h5))
    loudness = float(get_loudness(h5))
    mode = int(get_mode(h5))
    mode_confidence = float(get_mode_confidence(h5))
    song_hotttnesss = float(get_song_hotttnesss(h5))
    song_id = get_song_id(h5).decode('UTF-8')
    tempo = float(get_tempo(h5))
    time_signature = int(get_time_signature(h5))
    time_signature_confidence = float(get_time_signature_confidence(h5))
    title = get_title(h5).decode('UTF-8')
    track_id = get_track_id(h5).decode('UTF-8')
    year = int(get_year(h5))
    
    return [artist_name, artist_familarity, artist_hotttnesss, \
            artist_id, artist_location, danceability, duration, \
            energy, loudness, mode, mode_confidence, song_hotttnesss, \
            song_id, tempo, time_signature, time_signature_confidence, title, track_id, year]
```
